```python
import pandas as pd
```

```python
df_emp = pd.read_csv("emp.csv")
```

```python
df_sal = pd.read_csv("salary.csv")
```

```python
# Left outer join
# SQL : select * from emp as e left join sal as s on e.eeno = s.eno
df_emp.merge(df_sal,left_on = "eeno",right_on = "eno", how = "left")
```

```python
# Right outer join ( above left joins equivalent )

# SQL : select * from emp as e left join sal as s on e.eeno = s.eno
#df_emp.merge(df_sal,left_on = "eeno",right_on = "eno", how = "left")

# SQL : select * from sal as s right join emp as e on e.eeno = s.eno
df_sal.merge(df_emp,right_on = "eeno",left_on = "eno", how = "right")
```

```python
# SQL : select * from emp as e right join sal as s on e.eeno = s.eno
df_emp.merge(df_sal,left_on = "eeno",right_on = "eno", how = "right")
```

```python
# SQL : select * from emp as e full outer join sal as s on e.eeno = s.eno
df_emp.merge(df_sal,left_on = "eeno",right_on = "eno", how = "outer")
```

```python
# SQL : select * from sal as s full outer join emp as e on e.eeno = s.eno
df_outer = df_sal.merge(df_emp,left_on = "eno",right_on = "eeno", how = "outer")
```

```python
# write the output in a csv file in disk ( machine location)
df_outer.to_csv("Outer_joins_resultSet.csv",index=False)
```
