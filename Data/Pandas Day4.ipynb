```python
import pandas as pd
```

```python
df = pd.read_csv("emp_details.csv")
```

```python
# Q1 . Get me employees count in each department ans store the output in a new dataframe
# SQL : select deptno,count(*) from emp group by deptno
df_count_eno = df.groupby('deptno')['eno'].count().reset_index().rename(columns={"eno":"emp_count"})
```

```python
# Group the data based on multiple columns
df_count_eno_sal = df.groupby(['deptno','salary'])['eno'].count().reset_index().rename(columns={"eno":"emp_count"})
```

```python
df_count_eno_sal
```

```python
# perform multiple agrregation ( sum,min,max,count)
# Get me minimun,maximun,total salary,count of employees for each department
# SQL : select min(salary),max(salary),sum(salary),count(*) from emp group by deptno
df_multi_aggr = df.groupby('deptno')['salary'].agg(['min','max','sum','count']).reset_index()
```

```python
mapped_coumns = {"min":"minimum_salary","max":"maximum_salary","sum":"total_salary","count":"total_employees_count"}
df_multi_aggr.rename(columns = mapped_coumns)
```

```python
# merge function is used to join 2 dataframes . Its equivalent to joins in SQL ( inner , outer join ( left and right )))
df_sal = pd.read_csv('salary.csv')
```

```python
df_emp = pd.read_csv('emp.csv')
```

```python
df_emp
```

```python
df_sal
```

```python
# Get me eno,sal,ename by joining these dataframes
# SQL : select e.eno,e.ename,s.salary from emp as e inner join sal as s on e.eno = s.eno
df_emp.merge(df_sal,on='eno',how='inner')
```

```python
# Get me eno,sal,ename by joining these dataframes in case joining columns have different names
# SQL : select e.eno,e.ename,s.salary from emp as e inner join sal as s on e.eeno = s.eno
df_ans = df_emp.merge(df_sal,left_on='eeno',right_on='eno',how='inner')
required_column_list = ['eno','ename','salary']
df_ans[required_column_list]
```
